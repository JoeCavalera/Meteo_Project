```python
# Svuoto tutte le tabelle dati del progetto prima del run completo.
catalogo = "progetto_meteo"

spark.sql(f"USE CATALOG {catalogo}")


# Qui elenco solo le tabelle dati da svuotare.
# Non svuoto le tabelle metadati perché servono per città e zone ENTSO-E.
tabelle_da_svuotare = [
    f"{catalogo}.bronze.meteo_storico",
    f"{catalogo}.bronze.meteo_streaming",
    f"{catalogo}.bronze.dati_entsoe",

    f"{catalogo}.silver.dati_aggiornati",
    f"{catalogo}.silver.energy_block",

    f"{catalogo}.gold.dati_aggregati",
    f"{catalogo}.gold.dati_studio"
]


# Qui svuoto tutte le tabelle dati.
for tabella in tabelle_da_svuotare:
    spark.sql(f"TRUNCATE TABLE {tabella}")
    print(f"Svuotata: {tabella}")


# Controllo finale delle righe rimaste.
for tabella in tabelle_da_svuotare:
    righe = spark.table(tabella).count()
    print(f"{tabella}: {righe} righe")


print("Reset completo terminato correttamente.")
```
